# 49. Group Anagrams
https://leetcode.com/problems/group-anagrams/



```python
# Input
strs = ["eat", "tea", "tan", "ate", "nat", "bat"]

# Expected Output
# [["bat"],["nat","tan"],["ate","eat","tea"]]
```

```python
import collections

anagrams = collections.defaultdict(list)
anagrams
```

```python
for word in strs:
    anagrams[''.join(sorted(word))].append(word)
anagrams
```
